***Cleaning Data***

```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
print(df)
```

```python
print(df.head())
```

```python
print(df.tail())
```

```python
df.isna().sum()
```

```python
print(df["Calories"].to_string())
```

```python
new_df = df.dropna()
print(new_df.to_string())
```

```python
mean_calories = df["Calories"].mean()
print(mean_calories)
```

```python
df["Calories"].fillna(mean_calories,inplace=True)
print(df["Calories"].to_string())
```

```python
df.isna().sum()
```

```python
median_calories = df["Calories"].median()
print(median_calories)
```

```python
df["Calories"].fillna(median_calories,inplace=True)
print(df["Calories"].to_string())
```

```python
df["Calories"].fillna(0,inplace=True)
print(df["Calories"].to_string())
```

```python
print(df.to_string())
```

```python
df.loc[60,'Duration'] = 55
```
