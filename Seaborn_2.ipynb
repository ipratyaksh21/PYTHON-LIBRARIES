```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset("penguins")
df
```

```python
for cols in df.describe(include='object'):
    print(cols)
    print(df[cols].unique())
    print(" ")
```

## Lineplot

```python
ax1 = sns.lineplot(x="bill_length_mm",y="bill_depth_mm",hue="island",style="island",data=df,palette='winter')
plt.figure(figsize=(6,5))
plt.show(ax1)
```

```python
ax1 = sns.lineplot(x="bill_length_mm",y="flipper_length_mm",hue="sex",data=df,palette='rocket',markers=["o","*"],legend=False)
plt.grid()
plt.title("Bill len vs Flipper len",fontsize=15)
plt.xlabel("Bill lengeth (mm)")
plt.ylabel("Flipper lengeth (mm)")
plt.show(ax1)
```

```python
ax1 = sns.lineplot(x="body_mass_g",y="flipper_length_mm",style="sex",hue="species",data=df,palette='rocket')
plt.grid()
plt.title("Body Mass vs Flipper len",fontsize=15)
plt.xlabel("Body Mass (g)")
plt.ylabel("Flipper lengeth (mm)")
plt.show(ax1)
```

# Bar Plot

```python
bar = sns.barplot( x=df.island, y=df.body_mass_g, hue="sex",data =df, palette ='muted',saturation=1)
plt.title("Island vs Body Mass")
plt.xlabel("Island")
plt.ylabel("Body Mass (g)")
plt.show(bar)
```

```python
bar = sns.barplot( x="species", y="bill_length_mm", hue="sex",data =df, palette ='winter')
plt.title("species vs Bill Length")
plt.xlabel("species")
plt.ylabel("Bill Length")
plt.show(bar)
```

```python
order_1 = ["Dream","Torgersen","Bisco Island"]
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex",order= order_1,hue_order=["Female","Male"])
plt.show()
```

```python
bar1 = sns.barplot(x="bill_depth_mm",y="bill_length_mm",data=df,orient="h")
plt.show(bar1)
```

```python
sns.set(style="darkgrid")
sns.barplot(x="island",y="bill_length_mm",data=df,hue="sex",palette="winter",errcolor="r",errwidth=5)
plt.show()
```

# Histogram

```python
sns.set(style="whitegrid")
sns.displot(x="flipper_length_mm",data=df,color='k',alpha=0.8,kde =True)
plt.show()
```

```python
sns.set(style="darkgrid")
sns.displot(x="flipper_length_mm",data=df,color='Purple',alpha=0.8,kde =True,bins=[170,180,190,200,210,220,230,240])
plt.show()
```

# Scatter Plot

```python
sns.set(style="whitegrid")
mypalette = {"Male":'r',"Female":"k"}
sc =sns.scatterplot(x="bill_depth_mm",y="bill_length_mm",hue="sex",style="sex",data=df,palette=mypalette,size="sex")
plt.show(sc)
```

# Violin Plot

```python
print(tips.to_string())
```

```python
# Q: Visualization: Finding Highest Tip on Days by a sex.

tips = sns.load_dataset("tips")
ax1 = sns.violinplot(x="day",y="tip",hue="sex",data=tips,split =True,saturation=1)
plt.show(ax1)

# A: Male paid high tip on Saturday.
```

```python
# Q: Visualization: Finding Highest total_bill on Days by a Time.

tips = sns.load_dataset("tips")
ax1 = sns.violinplot(x="day",y="total_bill",hue="time",data=tips,linewidth=2,palette="flare")
plt.show(ax1)

# A:  Saturday at Dinner Time
```

```python
# Find a Best time to earn high tip on which day.

tips = sns.load_dataset("tips")
ax1 = sns.violinplot(x="time",y="tip",hue="day",data=tips,linewidth=1,palette="Accent",saturation=2)
plt.show(ax1)

# Best Chances are at Dinner Time and Saturday.
```
