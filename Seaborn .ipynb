```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# ***Using Tips Dataset***

```python
tips = sns.load_dataset("tips")
print(tips.head())
```

## ***Count Plot***

```python
tips.info()
```

```python
tips.describe(include ='category')
```

```python
for cols in tips.describe(include ='category'):
    print(cols)
    print(tips[cols].unique())
    print(" ")
    
```

```python
plt.figure(figsize=(6,5))
plt.title("Gender vs Smoking Habits")
palette_colors = {"Yes":"r","No":"g"}
sns.countplot(x="sex",hue="smoker",data=tips,palette=palette_colors,edgecolor='k')
plt.show()
```

```python
plt.figure(figsize=(6,5))
plt.title("Day vs Smoking Habits")
palette_colors = {"Yes":"r","No":"g"}
sns.countplot(x="day",hue="smoker",data=tips,palette=palette_colors,edgecolor='k')
plt.show()
```

```python
plt.figure(figsize=(6,5))
plt.title("Day vs Smoking Habits")
palette_colors = {"Lunch":"pink","Dinner":"purple"}
sns.countplot(x="day",hue="time",data=tips,palette=palette_colors,edgecolor='k')
plt.show()
```

```python
sns.histplot(data=tips, x="day",color="Purple")
```

## ***Scatter Plot***

```python
pallete_color = {"Lunch":"r","Dinner":"k"}
sns.scatterplot(x="tip",y="day",hue="time",palette= pallete_color,data=tips)
plt.title("Tips vs Day vs Time")
plt.show()
```

```python
pallete_color = {"Lunch":"r","Dinner":"k"}
sns.scatterplot(x="total_bill",y="tip",hue="time",palette= pallete_color,data=tips)
plt.title("Tips vs Day vs Time")
plt.show()
```

```python
# pallete_color = {"Lunch":"r","Dinner":"k"}
sns.lineplot(x="day",y="tip",hue="sex",data=tips)
plt.title("Tips vs Day vs Time")
plt.show()
```

## ***Line Plot***

```python
df = sns.load_dataset('fmri')
df
```

```python
df.info()
```

```python
df.columns
```

```python
ash = plt.figure(figsize=(6,5))
ash = sns.lineplot(x="timepoint",y="signal",data=df,color="Purple")
plt.xlabel("Timepoint")
plt.ylabel("Signals")
ash = sns.set_style("whitegrid")
plt.show(ash)
```

```python
df.describe()
```

```python
ash = plt.figure(figsize=(6,5))
ash = sns.lineplot(x="timepoint",y="signal",hue ="event",style="region",data=df,color="Purple",markers=True)
plt.xlabel("Timepoint")
plt.ylabel("Signals")
ash = sns.set_style("whitegrid")
plt.show(ash)
```

```python
df.describe(include='object')
```

```python
for cols in df.describe(include='object'):
    print(cols)
    print(df[cols].unique())
```

# ***Using Bar Plot***

```python
df = sns.load_dataset("titanic")
print(df.head())
```

```python
df.info()
```

```python
ax1 = sns.barplot(x="survived",y="age",hue="sex",data=df,edgecolor="k")
plt.title("Survival by Gender and Class")
plt.show(ax1)
```

```python
ax1 = sns.barplot(x="sex",y="survived",hue="class",data=df,edgecolor="k")
plt.title("Survival by Gender and Class")
plt.show(ax1)
```

```python
ax1 = sns.barplot(x="embark_town",y="survived",hue="who",data=df,edgecolor="k")
plt.title("Survived by Town with Who")
plt.show(ax1)
```
